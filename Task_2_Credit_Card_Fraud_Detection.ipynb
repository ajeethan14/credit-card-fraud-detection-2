```python
from tkinter import *
import joblib
import os


def load_model():
    model_path = "credit_card_model"
    if os.path.exists(model_path):
        return joblib.load(model_path)
    else:
        raise FileNotFoundError(f"Model file '{model_path}' not found.")


def show_entry_fields():
    try:
        values = [float(e.get()) for e in entry_fields]

        model = load_model()
        y_pred = model.predict([values])

        result = "Normal Transaction" if y_pred[0] == 0 else "Fraudulent Transaction"

        print("######################################")
        print("Credit Card Fraud Detection System:", result)
        print("######################################")

        Label(
            master, text="Final Prediction From the Model - Credit Card Fraud Detection"
        ).grid(row=32, columnspan=2)
        Label(master, text=result, fg="green" if y_pred[0] == 0 else "red").grid(
            row=33, columnspan=2
        )
    except ValueError:
        Label(
            master, text="Error: Please enter valid numerical values for all fields."
        ).grid(row=32, columnspan=2)
    except FileNotFoundError as e:
        Label(master, text=f"Error: {str(e)}").grid(row=32, columnspan=2)
    except Exception as e:
        Label(master, text=f"An unexpected error occurred: {str(e)}").grid(
            row=32, columnspan=2
        )


master = Tk()
master.title("Credit Card Fraud Detection System")

Label(
    master, text="Credit Card Fraud Detection System", bg="black", fg="white", width=30
).grid(row=0, columnspan=2)

entry_fields = []
for i in range(1, 31):
    Label(master, text=f"Enter value of V{i}").grid(row=i)
    entry = Entry(master)
    entry.grid(row=i, column=1)
    entry_fields.append(entry)

Button(master, text="Predict", command=show_entry_fields).grid(
    row=31, column=1, sticky=W, pady=4
)

mainloop()
```
